# EDA Profesional – Examen Parcial

Análisis exploratorio descriptivo del dataset `dataset_parcial_features.parquet`.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_parquet('data/processed/dataset_parcial_features.parquet')
display(df.head())
display(df.describe(include='all'))
```

```python
sns.countplot(x='y_riesgo', data=df)
plt.title('Distribución de la variable objetivo (y_riesgo)')
plt.show()
```

```python
num_cols = df.select_dtypes(include='number').columns
for col in num_cols:
    plt.figure(figsize=(5, 3))
    sns.histplot(df[col], kde=True)
    plt.title(f'Distribución de {col}')
    plt.tight_layout()
    plt.show()
```

```python
cat_cols = df.select_dtypes(include='object').columns
for col in cat_cols:
    plt.figure(figsize=(6, 3))
    df[col].value_counts().head(20).plot(kind='bar')
    plt.title(f'Frecuencias de {col} (top 20)')
    plt.tight_layout()
    plt.show()
```
