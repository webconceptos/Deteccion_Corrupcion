```python
from pathlib import Path
import markdown
from IPython.display import HTML, display

import subprocess
import shutil
```

```python
# Subimos dos niveles desde notebooks/Sprint4/
MD_PATH = Path("../../models/sprint4/reportes/informe_sprint4.md")
HTML_OUTPUT = Path("../../models/sprint4/reportes/informe_sprint4.html")
PDF_OUTPUT = Path("../../models/sprint4/reportes/informe_sprint4.pdf")

MD_PATH, HTML_OUTPUT, PDF_OUTPUT

```

```python
if not MD_PATH.exists():
    raise FileNotFoundError(f"No existe el informe Markdown: {MD_PATH}")

md_text = MD_PATH.read_text(encoding="utf-8")

print("Vista preliminar del informe:")
print(md_text[:800], "...")
```

```python
html = markdown.markdown(md_text, extensions=['tables', 'fenced_code'])

display(HTML(html))
```

```python
HTML_OUTPUT.write_text(html, encoding="utf-8")

print(f"HTML guardado en: {HTML_OUTPUT}")
```

```python
wkhtml = shutil.which("wkhtmltopdf")

if wkhtml:
    print(f"wkhtmltopdf detectado en: {wkhtml}")
else:
    print("‚ö†Ô∏è wkhtmltopdf NO est√° instalado.")
    print("   Descarga: https://wkhtmltopdf.org/downloads.html")
```

```python
if wkhtml:
    cmd = [
        wkhtml,
        str(HTML_OUTPUT),
        str(PDF_OUTPUT)
    ]
    subprocess.run(cmd, check=True)
    print(f"PDF generado correctamente en: {PDF_OUTPUT}")
else:
    print("‚ùå No se pudo generar PDF porque wkhtmltopdf no est√° instalado.")
```

```python
IMAGES_DIR = Path("../../models/sprint4")

images = [
    IMAGES_DIR / "shap/summary_plot.png",
    IMAGES_DIR / "shap/bar_plot.png",
    IMAGES_DIR / "reportes/ranking_f1.png",
    IMAGES_DIR / "pitch/pitch_modelos_f1.png"
]

valid_images = [img for img in images if img.exists()]

valid_images
```

```python
from IPython.display import Image

for img in valid_images:
    print(f"\nMostrando: {img}")
    display(Image(filename=str(img)))
```

```python
html_img = html

for img in valid_images:
    html_img += f'<br><img src="{img}" width="700"><br>'

HTML_OUTPUT.write_text(html_img, encoding="utf-8")

print(f"HTML con im√°genes guardado en: {HTML_OUTPUT}")
```

```python
if wkhtml:
    cmd = [
        wkhtml,
        str(HTML_OUTPUT),
        str(PDF_OUTPUT)
    ]
    subprocess.run(cmd, check=True)
    print(f"PDF generado correctamente con im√°genes en: {PDF_OUTPUT}")
else:
    print("‚ö†Ô∏è wkhtmltopdf no est√° instalado, PDF enriquecido no generado.")
```
