```python
# Sprint 4 – Notebook 00: Preparación Dataset

import pandas as pd
import numpy as np
from pathlib import Path

DATA_PATH = Path("../data/processed/dataset_modelado.parquet")
OUTPUT_PATH = Path("../models/sprint4/dataset_preparado.parquet")

df = pd.read_parquet(DATA_PATH)

df = df.fillna(0).replace([np.inf, -np.inf], 0)

TARGET = "riesgo_corrupcion"
y = df[TARGET]
X = df.drop(columns=[TARGET])

OUTPUT_PATH.parent.mkdir(parents=True, exist_ok=True)
df.to_parquet(OUTPUT_PATH)

print("Dataset preparado guardado en:", OUTPUT_PATH)
```
