# Validación de Resultados – Examen Parcial

Este notebook carga la tabla de resultados y genera gráficos comparativos de F1 y Recall por modelo.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# ======================================================
# RUTA CORRECTA PARA NOTEBOOKS Y PAPERMILL
# ======================================================
ROOT = os.getcwd()   # ← ESTA es la clave

def path(*args):
    return os.path.join(ROOT, *args)

# ======================================================
# Cargar resultados reales
# ======================================================
df = pd.read_csv(path("reports", "tablas", "resultados_modelos_parcial.csv"))

display(df)

plt.figure(figsize=(6, 4))
sns.barplot(data=df, x='modelo', y='f1')
plt.title('F1 promedio por modelo')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

plt.figure(figsize=(6, 4))
sns.barplot(data=df, x='modelo', y='recall')
plt.title('Recall promedio por modelo')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
