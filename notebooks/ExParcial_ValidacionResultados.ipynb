# Validación de Resultados – Examen Parcial

Este notebook carga la tabla de resultados y genera gráficos comparativos de F1 y Recall por modelo.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# ======================================================
# Patch rutas absolutas (funciona con papermill / jobs)
# ======================================================
import os

# Ruta absoluta a la raiz del proyecto (carpeta donde esta run_ExParcial_fulltareas.ps1)
ROOT = os.path.dirname(os.path.dirname(os.path.abspath("__file__")))

def path(*args):
    """Construye rutas absolutas seguras dentro del proyecto."""
    return os.path.join(ROOT, *args)


df = pd.read_csv(path("reports", "tablas", "resultados_modelos_parcial.csv"))

display(df)

plt.figure(figsize=(6, 4))
sns.barplot(data=df, x='modelo', y='f1')
plt.title('F1 promedio por modelo')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

plt.figure(figsize=(6, 4))
sns.barplot(data=df, x='modelo', y='recall')
plt.title('Recall promedio por modelo')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
